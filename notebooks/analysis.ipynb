```python
import pandas as pd
import plotly.express as px
```

```python
# We load the data from a CSV file
df = pd.read_csv('../data/dataset.csv', delimiter=';')
```

```python
# We clean the column names
df.columns = df.columns.str.strip()

# We fix some spelling errors in city names
df = df.replace({'Pachac�mac': 'Pachacámac', 'Mi Per�': 'Mi Perú'}, regex=True)

# We convert columns to numbers
df['Area'] = pd.to_numeric(df['Area'], errors='coerce')
df['Valor'] = pd.to_numeric(df['Valor'], errors='coerce')
df['Mantenimiento'] = pd.to_numeric(df['Mantenimiento'], errors='coerce')
df['Impuesto'] = pd.to_numeric(df['Impuesto'], errors='coerce')
df['Valor m2'] = pd.to_numeric(df['Valor m2'], errors='coerce')
```

```python
# We create a histogram to show the value distribution
fig = px.histogram(df, x='Valor', title='Distribucion del Valor')
fig.show()
```

```python
fig = px.scatter(df, x='Area', y='Valor', title='Area vs Valor')
fig.show()
```
